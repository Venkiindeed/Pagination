```python
# Data
data = [
    {"id":1,"name":"Product A","category":"Electronics","status":"active"},
    {"id":2,"name":"Product B","category":"Clothing","status":"inactive"},
    {"id":3,"name":"Product C","category":"Electronics","status":"active"},
    {"id":4,"name":"Product D","category":"Electronics","status":"inactive"},
    {"id":5,"name":"Product E","category":"Clothing","status":"active"},
    {"id":6,"name":"Product F","category":"Electronics","status":"active"}
]

def paginate(data, page, page_size, filters, sort_key, order):

    filtered = []

    # Step 1: Filtering
    for item in data:
        if item["category"] == filters["category"] and item["status"] == filters["status"]:
            filtered.append(item)

    # Step 2: Sorting
    if order == "asc":
        filtered.sort(key=lambda x: x[sort_key])
    else:
        filtered.sort(key=lambda x: x[sort_key], reverse=True)

    # Step 3: Pagination
    start = (page - 1) * page_size
    end = start + page_size

    return filtered[start:end]


# Inputs
page = 1
page_size = 3
filters = {"category":"Electronics", "status":"active"}

sort_key = input("Enter key to sort (id/name/category/status): ")
order = input("Enter order (asc/desc): ")

# Function call
result = paginate(data, page, page_size, filters, sort_key, order)

# Output line by line
for item in result:
    print(item)
```
